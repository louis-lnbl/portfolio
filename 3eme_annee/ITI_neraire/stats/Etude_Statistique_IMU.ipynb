```python
import pandas as pd
import numpy as np
```

```python
data_IMU_Statique = pd.read_csv("../data/Gr11/donnees/acquisitions_accelerometre_exterieur_statique.csv")
```

```python
analyse_data_IMU_Statique = data_IMU_Statique.describe()
print(analyse_data_IMU_Statique)
```

```python
data_IMU_Statique_accelx = data_IMU_Statique[["Accel X"]]
data_IMU_Statique_accely = data_IMU_Statique[["Accel Y"]]
data_IMU_Statique_accelz = data_IMU_Statique[["Accel Z"]]
data_IMU_Statique_gyrox = data_IMU_Statique[["Gyro X"]]
data_IMU_Statique_gyroy = data_IMU_Statique[["Gyro Y"]]
data_IMU_Statique_gyroz = data_IMU_Statique[["Gyro Z"]]


import matplotlib.pyplot as plt
plt.figure(figsize=(10,20))
plt.subplot(3,1,1)
plt.plot(data_IMU_Statique_accelx)
plt.xlabel("numro chantillon")
plt.ylabel("Acceleration en (m/s-2)")
plt.title("accel X")
plt.grid(True, linestyle="--", alpha=0.5)


plt.subplot(3,1,2)
plt.plot(data_IMU_Statique_accely)
plt.xlabel("numro chantillon")
plt.ylabel("Acceleration en (m/s-2)")
plt.title("accel Y")
plt.grid(True, linestyle="--", alpha=0.5)

plt.subplot(3,1,3)
plt.plot(data_IMU_Statique_accelz)
plt.xlabel("numro chantillon")
plt.ylabel("Acceleration en (m/s-2)")
plt.title("accel Z")
plt.grid(True, linestyle="--", alpha=0.5)
```

```python
plt.figure(figsize=(10,20))
plt.subplot(3,1,1)
plt.plot(data_IMU_Statique_gyrox)
plt.subplot(3,1,2)
plt.plot(data_IMU_Statique_gyroy)
plt.subplot(3,1,3)
plt.plot(data_IMU_Statique_gyroz)
```
